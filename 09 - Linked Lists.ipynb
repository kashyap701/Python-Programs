#Linked list

A linked list is a linear collection of data elements, whose order is not given by their physical placement in memory. Instead, each element points to the next. It is a data structure consisting of a collection of nodes which together represent a sequence.

##Singly linked list

A singly linked list is the one in which a node points to the object of the next node.

###Implementation of a singly linked list

Creating the node class

It has 2 data members namely data and next.

Data   :    For storing the data.
Next   :    For storing the next object.

Creating a Linked list class 

This class contains all the operations that we are going to perform on a linked list.

##Doubly Linked List

This is a variation of a linked list in which we can iterate backwards.

```
class Node():
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None


class DoublyLinkedList():
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def traverse(self, index):
        curNode = self.head
        for _ in range(index):
            curNode = curNode.next
            
        return curNode
        
    def InsertAtEnd(self, value):
        newNode = Node(value)
        
        if self.head == None:
            self.head = newNode
            self.tail = self.head
            self.length += 1
        
        else:            
            self.tail.next = newNode
            newNode.prev = self.tail
            self.tail = newNode
            self.length += 1

    def InsertAtFirst(self,value):
        newNode = Node(value)
        newNode.next = self.head
        self.head.prev = newNode
        self.head = newNode
        self.length += 1

    

    def InsertInMiddle(self, index, value):
        newNode = Node(value)
        if index==0:
            self.InsertAtFirst(value)
            return
        elif index>=self.length:
            self.InsertAtEnd(value)
            return
        else:
            leaderNode = self.traverse(index-1)
            holdNode = leaderNode.next
            
            leaderNode.next = newNode
            newNode.prev = leaderNode
            newNode.next = holdNode
            holdNode.prev = newNode
            self.length += 1

    def delete(self,index):
        if index==self.length-1:
            self.tail = self.tail.prev
            self.tail.next = None
            self.length -= 1
            return
        
        leaderNode = self.traverse(index-1)
        delNode = leaderNode.next
        holdNode = delNode.next
        leaderNode.next = holdNode
        holdNode.prev = leaderNode
        self.length -= 1
        

    def printList(self):
        curNode = self.head
        arr = []
        for _ in range(self.length):
            print(curNode.value, end="  ")
            curNode = curNode.next
        print()
```
